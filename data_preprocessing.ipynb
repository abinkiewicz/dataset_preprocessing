Obróbka danych
* Twoim zadaniem jest wykonanie obróbki danych zawartych w załączonym pliku CSV.
* Pozbądź się wszystkich rzędów które zawierają duplikaty
* Rozbij kolumnę `'Name'` na dwie kolumny `'Imie'` i `'Nazwisko'`
* Usuń kolumnę `'Name'`
* Napraw rzędy które zawierają brakuące dane
* Zmień nazwy kolumn:
    - `'Age'` -> `'Wiek'`
    - `'Height'` -> `'Wzrost'`
    - `'Weight'` -> `'Waga'`
    - `'Address'` -> `'Adres'`
      
* Dodaj kolumnę 'BMI' obliczoną jako `Waga / (Wzrost / 100) ^ 2`
* Posortuj DataFrame po kolumnie 'BMI' malejąco
* Zapisz zmodyfikowany DataFrame do pliku CSV
* Wykorzystaj nowo nabytą wiedzę z biblioteki Pandas
* Prześlij na discordzie (`#modul-4-zad3`) notebook z rozwiązaniem oraz plik CSV z wynikami


```python
import pandas as pd
```

```python
df = pd.read_csv('27__dane.csv', sep=';')

df
```

#### 1. Usunięcie duplikatów

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
df
```

#### 2. Rozbicie kolumny "name"

```python
df[['Imie', 'Nazwisko']] = df['Name'].str.split(' ', expand = True)
df
```

#### 3. Usunięcie kolumny "name"

```python
df = df.drop('Name', axis = 1)
df
```

#### 4. Uzupełnienie brakujących danych

```python
df['Height'] = df['Height'].fillna(df['Height'].mean())
df['Weight'] = df['Weight'].fillna(df['Weight'].mean())
df['Age'] = df['Age'].fillna(df['Age'].median())
df.isnull().sum()
```

Pozostałych danych, tj. Imię, Nazwisko i adres nie sposób wypełnić danymi bez wiedzy.

#### 5. Zmiana nazw kolumn:
* Age -> Wiek
* Height -> Wzrost
* Weight -> Waga
* Address -> Adres

```python
df = df.rename(columns = {'Age':'Wiek', 'Height':'Wzrost', 'Weight':'Waga', 'Address':'Adres'})
df
```

#### 6. Dodanie kolumny 'BMI" wg wzoru:  `Waga / (Wzrost / 100) ^ 2`

```python
df['BMI'] = df['Waga'] / (df['Wzrost'] / 100) ** 2
df
```

#### 7. Sortowanie DataFrame po kolumnie 'BMI' malejąco

```python
df = df.sort_values('BMI', ascending = False)
df
```

#### 8. Zapisanie zmodyfikowanego DataFrame do pliku CSV


```python
df.to_csv('27_dane_processed.csv', index = False)
```
